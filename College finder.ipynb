```python
import pandas as pd
import os

os.
k = pd.read_csv('mark2.csv')

mark = []
community = []
c_code = []
br_code = []

for i in k['S NO']:
    result = i.split(' ')[-6:]
    mark.append(result[0])
    community.append(result[2])
    c_code.append(result[3])
    br_code.append(result[4])
    
data = pd.DataFrame()
data['mark'] = mark
data['community'] = community
data['branch'] = br_code
data['college code'] = c_code


from bs4 import BeautifulSoup as bs
import requests as r
link = input('Enter the link:')
k = r.get(link)

soup = bs(k.content)

a = []
b = []
c = []
for i in soup.find_all('tr'):
    a.append(i)
for j in a:
    b.append(j.find_all('td'))
for k in b:
    for v in k:
        c.append(v.find_all('p'))
        
tex = []

for vv in c:
    for vvv in vv:
        tex.append(vvv.text)
        
codes = []
total = int(input('No of colleges:'))
total = (total * 4) + 5
st = input('Enter the starting number')
for i in tex[5:total]:
    if i[0] == st:
        codes.append(i)
        
a = 0 
code = codes
for i in data['college code']:
    for j in code:
        if i == j and data['community'][a] == 'BC':
            print(data['mark'][a],'--',data['college code'][a],'--',data['community'][a],'--',data['branch'][a])
    a+=1
```
